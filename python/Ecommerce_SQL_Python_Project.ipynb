# E-Commerce Data Analysis Using SQL & Python

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Import

```python
# Load CSV files
monthly = pd.read_csv("/content/1.Calculate the number of orders per month in 2018.csv")
seller = pd.read_csv("/content/5.Calculate the total revenue generated by each seller , and rank them by revenue.csv")
yearly = pd.read_csv("/content/Yearly_Sales.csv")
price = pd.read_csv("/content/4.Identify the correlation between product price and the number of times a product has been purchased.csv")
category = pd.read_csv("/content/3.Find the total sales per category.csv")
```

## Monthly Orders Trend

```python
# Orders per month 2018
plt.figure()
plt.plot(monthly.iloc[:,0], monthly.iloc[:,1])
plt.title("Orders Per Month - 2018")
plt.xlabel("Month")
plt.ylabel("Total Orders")
plt.show()
```

## Seller Performance

```python
# Top 10 Sellers
top10 = seller.head(10)

plt.figure()
plt.bar(top10.iloc[:,0], top10.iloc[:,1])
plt.title("Top 10 Sellers by Revenue")
plt.xticks(rotation=90)
plt.show()
```

## Yearly Sales Growth

```python
# Yearly Sales
plt.figure()
plt.plot(yearly.iloc[:,0], yearly.iloc[:,1])
plt.title("Yearly Total Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

## Price vs Demand Correlation

```python
plt.figure()
plt.scatter(price['avg_price'], price['purchase_count'])
plt.title("Price vs Purchase Frequency")
plt.xlabel("Average Price")
plt.ylabel("Purchase Count")
plt.show()
```

```python
# Calculate correlation value
corr = price['avg_price'].corr(price['purchase_count'])
print("Correlation:", round(corr,3))
```

## Category Revenue

```python
top10cat = category.head(10)

plt.figure()
plt.bar(top10cat['product_category'], top10cat['total_sales'])
plt.title("Top 10 Product Categories by Revenue")
plt.xticks(rotation=90)
plt.show()
```
